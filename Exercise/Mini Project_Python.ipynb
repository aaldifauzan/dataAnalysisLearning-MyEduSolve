### Nomor 1

```python
import pandas as pd
```

```python
def status(penghasilan):
    if penghasilan > 15:
        return "Crazy Rich"
    elif 5 <= penghasilan <= 15:
        return "Makmur"
    elif 2 <= penghasilan <= 5:
        return "Mandiri"
    elif 1 <= penghasilan <= 2:
        return "Butuh Pendampingan"
    elif 0 <= penghasilan <= 1:
        return "Rentan"
    else:
        return "Penghasilan tidak mungkin negatif"
```

```python
data = []

for i in range(10):
    input_data = input(f"Masukkan data keluarga ke-{i + 1} (Contoh: 'Nama | Penghasilan'): ")
    nama, penghasilan_str = input_data.split("|")
    nama = nama.strip()
    penghasilan = float(penghasilan_str.strip())
    
    if penghasilan < 0:
        print("Penghasilan tidak mungkin negatif. Data tidak disimpan.")
        continue
        
    status_ekonomi = status(penghasilan)
    data.append([nama, penghasilan, status_ekonomi])
```

```python
df = pd.DataFrame(data, columns=["Nama kepala keluarga", "Penghasilan bulanan dalam juta", "Status ekonomi"])

df.to_csv("data.csv", index=False)

print("Data berhasil disimpan.")
```

```python
import numpy as np
import pandas as pd
tugas = np.random.normal(85, 7, 40)
uts = np.random.normal(60, 15, 40)
uas = np.random.normal(75, 10, 40)
data = pd.DataFrame({'Tugas': tugas, 'UTS': uts, 'UAS': uas})
data['Nilai Akhir'] = 0.2 * data['Tugas'] + 0.4 * data['UTS'] + 0.4 * data['UAS']
def score(nilai):
    if nilai >= 85:
        return 'A'
    elif nilai >= 80:
        return 'A-'
    elif nilai >= 75:
        return 'B+'
    elif nilai >= 70:
        return 'B'
    elif nilai >= 65:
        return 'B-'
    elif nilai >= 60:
        return 'C+'
    elif nilai >= 55:
        return 'C'
    elif nilai >= 50:
        return 'C-'
    elif nilai >= 20:
        return 'D'
    else:
        return 'E'

data['Kategori Nilai'] = data['Nilai Akhir'].apply(score)

data.head(40)
```
