## **Descriptive Statistics - Basic**

**1. Introduction**

**What is Descriptive Statistics?**

Descriptive statistics is a branch of statistics that deals with the collection, analysis, interpretation, and presentation of data. Its primary goal is to describe the main features of numerical data in a quantitative manner without making any assumptions about the underlying population Descriptive statistics provide a summary of the main aspects of the data, such as the average value, distribution, and spread.

**Importance and Applications of Descriptive Statistics**

Descriptive statistics play a crucial role in the initial stage of data analysis. They provide a snapshot of the main characteristics of the data and offer a graphical representation that can help in understanding the data's underlying structure. Some applications include:
- Summarizing large amounts of data in a meaningful way
- Helping in understanding patterns in the data
- Assisting in making decisions.
- Providing a basis for further statistical analysis.

**2. Measures of Central Tendency**

Central tendency measures give us a central value of data distribution. They are vital as they summarize a large dataset with a single value that represents the center of its distribution

**Mean**: It's the average of all the data points, it's calculated by summing all data points and dividing by the number of data points.

**Median**: It's the middle value in a dataset when the values are arranged in ascending or descending order. If there's an even number of observations, the median is the average of the two middle numbers.

**Mode**: It's the value that appears most frequently in a dataset. A dataset can have one mode, more than one mode or no mode at all.

**Theory Behind Central Tendency**

**Mean**: While it gives a quick snapshot of the data, it's sensitive to outliers. For instance, adding a very high or very low value can drastically change the mean

**Median**: It's more robust than the mean. Since it's based on the middle values), cutliers don't affect it as much. It's especially useful for skewed distributions

**Mode**: It provides information about the most common values in the dataset. It's the only measure of central tendency that can be used with categorical data

**3. Measures of Dispersion**

Measures of dispersion tell us how spread out our data is. They provide insight into the variability of the dataset.

![](https://sixsigmastudyguide.com/wp-content/uploads/2020/10/di1.png)

**Range**: It's the difference between the maximum and minimum values in the dataset

**Variance**: It measures how far each number in the sets from the mean and is calculated by taking the average of the squared differences from the Mean.

**Standard Deviation**: It's the square root of the variance and measures the amount of variation or dispersion of a set of values.

**Interquartile Range (IQR)**: it's the range between the first quartile (25th percentile) and the third quartile (75th percentile). It gives us an idea about the middle 50% of scores

```python
for i in df:
    try:
        print(f'{i} range {df[i].max()-df[i].min()}')
    except:
        pass
```

**4. Measures of Shape**

These measures provide insight into the shape of the distribution of the data.
![](https://ars.els-cdn.com/content/image/3-s2.0-B9780128207178000087-f05-03-9780128207178.jpg)

**Skewness**: It measures the asymmetry of the distribution of values. A negative skew indicates a distribution that is skewed left, while a positive skew indicates a distribution that is skewed right.

**Kurtosis**: It measures the "tailedness" of the distribution. High kurtosis indicates a lot of outliers, and low kurtosis indicates a lack of outliers.

```python
# Computing the skewness for each feature
df.skew()
```

**6. Correlation and Covariance**

**Correlation**: It measures the strength and direction of the linear relationship between two variables, the values range between -1 and 1.

**Covariance**: It measures the directional relationship between two variables.

![](https://excelrcom.b-cdn.net/assets/admin/ckfinder/userfiles/images/2020%20uploads/blog1.PNG)

```python
# Computing the correlation matrix
df.corr()
```

```python
# Visualizing with correlation matrix
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
# Visualizing with pair plot
sns.pairplot(df)
plt.show()
```

```python
# Computing the covariance matrix
df.cov()
```

```python
# Visualizing correlation matrix
sns.heatmap(df.cov(),annot=True)
plt.show()
```

**Correlation doesn't imply Causation**
- However, correlation does not imply causation. There may be, for example, an unknown factor that influences both variables similarly
- Causation indicates that one event is the result of the occurrence of the other event: i.e. there is a causal relationship between the two events. This is also referred to as cause and effect.
- A statistically significant correlation has been reported, for example, between yellow cars and a lower incidence of accidents. That does not indicate that yellow cars are safer, but just that fewer yellow cars are involved in accidents. A third factor, such as the personality type of the purchaser of yellow cars, is more likely to be responsible than the color of the paint itself.
