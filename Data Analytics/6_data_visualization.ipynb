## Python Data Visualization

While working with data, it can be difficult to truly understand your data when it's just in tabular form. To understand what exactly our data conveys, and to better clean it and select suitable models for it, we need to visualize it or represent it in pictorial form. This helps expose patterns, correlation, and trends that cannot be obtained when data is in a table or CSV file.

The process of finding trends and correlations in our data by representing it pictorially is called Data Visualization. To perform data visualization in Python, we can use various Python data visualization modules such as Matplotlib, Seaborn, Plotly, etc. In this article, The Complete Guide to Data Visualization in Python, we will discuss how to work with some of these modules for data visualization in Python and cover the following topics in detail.

### What is Data Visualization?

Data Visualization is a field in data analysis that deals with visual representation of data. It graphically plots data and is an effective way to communicate inferences from data.

Using data visualization, we can get a visual summary of our data. With pictures, maps, and graphs, the human mind has an easier time processing and understanding any given data. Data visualization plays a significat role in thre representation of both small and large data sets, but it is especially useful when we have large data sets, in which it is impossible to see all of our data, let alon process and understand it manually.

Data Visualization with Pandas is the representation of data in a graphical format. It helps people understand the significance of data by summarizing and presenting a huge amount of data in a simple and easy-to-understand format and helps communicate information clearly and effectively.

```python
# Import Libraries

# Data wrangling
import numpy as np
import pandas as pd

# Data visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Import warning
import warnings
warnings.filterwarnings("ignore")
```

### 1. Line Charts

**What are Line Charts?**

Line charts are a type of graph that displays data points connected by staight lines.

They are commonly used for visualizing data that changes over time or has a natural sequence.

**When to Use Line Charts?**

Tracking trends and changes over time.

Showing relationships between continous variables.

```python
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"

data = pd.read_csv(url)
```

```python
# Display the first few rows of the datase
data.set_index(["Month"])
```

```python
data.info()
```

**Creating a Basic Line Chart**

To create a basing chart, you can use **Matplotlib** `plot` function.

But first we need to transform the timeseries data type into a proper timeseries.

```python
# Transform datatype with with pd.to_datetime
data["Month"] = pd.to_datetime(data["Month"])
```

```python
# Create a basic line chart for Air Passenger data
plt.figure(figsize=(20,5))
plt.plot(data['Month'], data['Passengers'],
        label='Passenger Count', color='c', marker='v', linestyle='--')
plt.title('International Airline Passengers Over Time')
plt.xlabel('Month')
plt.ylabel('Passenger Count')
plt.xticks(rotation=45, ha='center') # Rotate month labels for readibility
plt.legend()
plt.grid(linestyle='--', alpha=0.8)
plt.show()
```

**Seaborn** provides additional styling options and makes it easier to create visually appealing line charts by further customize line charts by adjusting colors, markers, labels, etc

```python
# Create a basic line chart for Air Passenger data
plt.figure(figsize=(20,5))
sns.lineplot(data=data, x='Month', y='Passengers', label='Passenger Count', color='green', marker='o',linestyle='--')
plt.title('International Airline Passengers Over Time')
plt.xlabel('Month')
plt.ylabel('Passenger Count')
plt.xticks(rotation=45, ha='center') # Rotate month labels for readibility
plt.legend()
plt.grid(linestyle='--', alpha=0.8)
plt.show()
```

### 2. Bar Graphs

**What are Bar Graphs?**

Bar graphs are charts that use rectangular bars to represent data.

They are suitable for visualizing and comparing categorical data

**When to Use Bar Graphs?**

Comparing data across different categories

Displaying data that doesn't have a natural order or sequence

```python
# Load the tips dataset
tips = sns.load_dataset('tips')
```

```python
tips.info()
```

```python
tips.head()
```

**Creating a Basic Vertical Bar Chart**

To create a basic vertical bar chart, use **Matplotlib** `bar()` function.

```python
# Create a basic vertical bar chart for day-wise tip totals
plt.figure(figsize=(7,4))
plt.bar(tips['day'], tips['tip'])
plt.title('Total Tips by Day')
plt.xlabel('Day')
plt.ylabel('Total Tips')
plt.grid(axis='y', linestyle='--',alpha=0.6)
plt.show()
```

```python
# Create a basic vertical bar chart using seaborn
plt.figure(figsize=(7,4))
sns.barplot(data=tips, x='day', y='tip', palette='Blues')
plt.title('Total Tips by Day')
plt.xlabel('Day')
plt.ylabel('Total Tips')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

Horizontal bar charts are useful when you have long category labels

```python
# Create a horizontal bar chart for gender-wise tip totals
plt.figure(figsize=(10,5))
sns.barplot(data=tips, y='sex', x='tip', palette='Greens')
plt.title('Total Tips by Gender')
plt.xlabel('Total Tips')
plt.ylabel('Gender')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```

Stacked bar charts are useful for showing the composition of categories.

Visualization with pandas (https://pandas.pydata.org/docs/user_guide/visualization.html)

```python
# Group data
time_tip_totals = tips.groupby('time')['tip'].sum().reset_index()
```

```python
# Create a stacked bar for time-wise tip totals
ax = time_tip_totals.plot(kind='bar', x='time', y='tip', stacked=True, figsize=(10,5), colormap='viridis')
plt.title("Total Tips by Time of Day")
plt.xlabel("Time of Day")
plt.ylabel("Total Tips")
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Add numerical labels above each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}',
                (p.get_x() + p.get_width() / 2.0, p.get_height()),
                textcoords='offset points', xytext=(0, 5),
                ha='center', fontsize=8, color='black')
plt.show()
```

**Creating a Basic Vertical Bar Chart with Numerical Labels**

To create a basic vertical bar chart with numerical labels, use **Seaborn** `barplot()` function.

```python
# Create a basic vertical bar chart  for day-wise tip total with numerical labels
plt.figure(figsize=(10,5))
ax = sns.barplot(data=tips, x='day', y='tip', palette='Blues')
plt.title("Total Tips by Time of Day")
plt.xlabel("Time of Day")
plt.ylabel("Total Tips")
plt.grid(axis='y', linestyle='--', alpha=0.6)
# Add numerical labels above each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}',
                (p.get_x() + p.get_width() / 1.5, p.get_height()),
                textcoords='offset points', xytext=(0, 10),
                ha='center', fontsize=12, color='black')
plt.show()
```

```python
# Create a horizontal bar chart for gender-wise tip totals with numerical labels
plt.figure(figsize=(10,5))
ax = sns.barplot(data=tips, y='sex', x='tip', palette='Greens')
plt.title("Total Tips by Gender")
plt.xlabel("Total Tips")
plt.ylabel("Gender")
plt.grid(axis='x', linestyle='--', alpha=0.6)
# Add numerical labels above each bar
for p in ax.patches:
    ax.annotate(f'{p.get_width():.2f}',
                (p.get_width(),  p.get_y()+ p.get_height() / 2.),
                textcoords='offset points', xytext=(15, 10),
                ha='center', va='center', fontsize=12, color='black')
plt.show()
```

### 3. Histograms

**What are Histogram?**

Histograms are graphical representations of the distribution of a dataset

They display the frequency (or count) of data points within specified intervals or "bins"

**When to Use Histograms?**

Visualizing the distribution of numerical data.

Identifying patterns, skewness, and central tendencies in data

**Creating a Histogram**

To create a histogram using **Matplotlib**, use the `hist()` function.

```python
# Create a histogram for the distribution of total bill amounts using Matplotlib
plt.figure(figsize=(10,5))
plt.hist(tips['total_bill'], bins=20, color='green', alpha=0.8, edgecolor='black')
plt.title('Distribution of Total Bill Amounts (Matplotlib)')
plt.xlabel('Total Bill Amount')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

We can also use **Seaborn** `histplot()` function for a better visualization.

You can customize histograms by adjusting the number of bins, adding kernel density estimation (KDE), and changing colors.

```python
# Create a histogram for the distribution of total bill amounts using Seaborn
plt.figure(figsize=(10,5))
ax = sns.histplot(data=tips, x='total_bill', bins=20, kde=True, color='skyblue', line_kws={'lw': 4, 'ls': '--'})
ax.lines[0].set_color('red')  # line color
plt.title('Distribution of Total Bill Amounts (Matplotlib)')
plt.xlabel('Total Bill Amount')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

### 4. Scatter Plot

**What are Scatter Plot?**

Scatter plots are graphs that display individual data points as dots on two-dimentional plane.

They are used to visualize the relationship between two numerical variables.

**When to Use Scatter Plot?**

Exploring relationships and correlations between variables.

Identifying patterns, clusters, or outliers in data.

**Creating a Scatter Plot**

To create a scatter plot using **Matplotlib**, use the `scatter()` function.

Matplotlib allows customization of scatter plot by adjusting colors, markers, and styles.

```python
# Customizing the Matplotlib scatter plot appearance
plt.figure(figsize=(10,5))
plt.scatter(tips['total_bill'], tips['tip'],
           color='green', marker='o', s=50,
           alpha=0.8, edgecolor='black', linewidth=1.2)
plt.title('Scatter Plot of Total Bill vs. Tip (Matplotlib)')
plt.xlabel('Total Bill Amount')
plt.ylabel('Tip Amount')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

**Seaborn** also provides options for customization of scatter plots, including style adjustments and color palettes.

```python
# Customizing the Seaborn scatter plot appearance
plt.figure(figsize=(10,5))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex',
           palette='viridis', size='tip', sizes=(20,200), alpha=0.7,
           edgecolor='black', linewidth=1.2)
plt.title('Scatter Plot of Total Bill vs. Tip (Seaborn)')
plt.xlabel('Total Bill Amount')
plt.ylabel('Tip Amount')
plt.grid(True, linestyle='--', alpha=0.8)
plt.legend(title='Gender', loc='upper right', bbox_to_anchor=(1.25,1))
plt.show()
```

```python
# Using Regplot
plt.figure(figsize=(8,4))
sns.regplot(data=tips, x='total_bill', y='tip', color=".3", line_kws=dict(color="r"))
plt.title("RegPlot of Total Bill vs. Tip (Seaborn)")
plt.xlabel("Total Bill Amount")
plt.ylabel("Tip Amount")
plt.grid(True, linestyle='--', alpha=0.8)
plt.show()
```

### 5. Heat Maps

**What are Heat Maps?**

Heat Maps are graphical representations of data where values are depicted as colors.

They are useful for visualizing matrices, correlations, and distributions.

**When to Use Heat Maps?**

Exploring relationships and patterns in large dataset.

Visualizing data with two dimensions, such as time-series data or correlation matrices.

**Creating a Heat Map**

To create a basic heat map using **Matplotlib**, you can use `imshow()` function to visualize a 2D matrix of data.

```python
# Create correlation matrix between numerical columns
tips_num = tips.select_dtypes(exclude='object')
corr_matrix = tips_num.corr()
```

```python
plt.figure(figsize=(8,6))
plt.imshow(corr_matrix, cmap='YlGnBu', interpolation='nearest', aspect='auto', vmin=-1, vmax=1)
plt.title('Customized Heat Map of Correlation (Matplotlib)')
plt.colorbar(label='Correlation Coefficient')
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=45)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.show()
```

**Seaborn** provides extensive customization options for heat maps, including annotations and color palettes.

```python
# Customizing the Seaborn heat map appearance
plt.figure(figsize=(7,7))
sns.heatmap(corr_matrix, annot=True, linewidths=1,
           linecolor='black', fmt='.2f', cbar_kws={'label': 'Correlation Coefficient'})
plt.title('Customized Heat Map of Correlations (Seaborn)')
plt.show()
```

### 6. Pie Chart

**What are Pie Chart?**

Pie Charts are useful for visualizing the composition of a whole and comparing parts to the whole.

**When to Use Pie Chart?**

Displaying the distribution of categorical data.

Showing the composition of a whole (e.g, budget allocation, meal expenses).

You can use **Matplotlib** `pie()` to create the Pie Chart

```python
# Calculate the sum of total bills for different days
bill_sum_by_day = tips.groupby('day')['total_bill'].sum()

# labels for each slice
labels = bill_sum_by_day.index
```

```python
# Create basic Pie Chart?
explode = (0.1, 0, 0, 0)
plt.pie(bill_sum_by_day, labels=labels, autopct='%1.1f%%', startangle=90, explode=explode)
plt.title('Exploded Pie Chart')
plt.legend(labels, loc='upper right')
plt.show()
```

You can also customize the Pie Chart using **Seaborn** style and color palette options

```python
# Set explode
explode = (0, 0, 0, 0.1)
```

```python
#  Display percentage and absolute values
plt.figure(figsize=(5,5))
sns.set_style("whitegrid")
sns.set_palette("pastel")
plt.pie(bill_sum_by_day, labels=labels, autopct='%1.1f%%', startangle=90, pctdistance=0.85, explode=explode)
plt.title('Pie Chart with Percentage and Absolute Values (Seaborn)')
plt.legend(labels, loc='best', bbox_to_anchor=(1.0,1.0))
plt.show()
```

### 8. Facet Grid

**What are Facet Grid?**

Facet Grid allow you to display and compare multiple visualizations within the same figure.

This technique is useful for showing related data or comparisons.

**Creating Facet Grid (Multiple Graphs in One Plot)**

You can use **Seaborn** Facet Grid to create multiple graphs in one plot.

```python
# Create multiple graphs (histograms) for total bill amounts aqnd tips in one plot
g = sns.FacetGrid(tips, col='sex', hue='time', palette='viridis')
g.map(sns.histplot, 'total_bill', bins=15, alpha=0.7)
g.add_legend(title='Time of Day')
g.set_axis_labels('Total Bill Amount', 'Frequency')
plt.subplots_adjust(top=0.8)
g.fig.suptitle('Multiple Histograms in One Plot (Seaborn)')
plt.show()
```

We can create a facet grid with four graphs consisting of scatter plots, histograms, and any other visualizations using **Seaborn** Facet Grid.

```python
# Create a FacetGrid with four graphs (2x2 grid)
g = sns.FacetGrid(tips, col='time', row='sex', height=3, aspect=1.5, margin_titles=True)

# Set the Plts
g.map_dataframe(sns.scatterplot, x='total_bill', y='tip', alpha=0.7) # Scatter plot (Total Bill vs. Tip)
g.map_dataframe(sns.histplot, x='total_bill',bins=15, color='skyblue', alpha=0.7) # Histogram of Total Bill
g.map_dataframe(sns.histplot, x='tip',bins=15, color='salmon', alpha=0.7, element='step') # Histogram of Tip

# Set Title and adjusting layout
g.set_axis_labels('Total Bill Amount', 'Tip Amount') # Customizing facet titles
g.set_titles(col_template="Time: {col_name}", row_template="Sex: {row_name}")
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Facet Grid with Multiple Graphs (Seaborn)')
plt.show()
```
